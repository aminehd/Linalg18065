```julia
using LinearAlgebra

# Create a nearly singular matrix
ε = 1e-10  # tiny number
A = [1.0   1.0;
     1.0   1.0+ε]  # Almost singular!

println("Nearly singular A:")
println(A)
println("Determinant = ", det(A), " (tiny!)")
println("Condition number = ", cond(A), " (huge!)")

println("\n" * "="^50 * "\n")

# Try to solve Ax = b
b = [2.0; 2.0]

# Method 1: Direct inversion (TERRIBLE!)
x_direct = inv(A) * b
println("Solution via A^(-1):")
println("x = ", x_direct)
println("||x|| = ", norm(x_direct), " ← HUGE!")

# Small perturbation in b
b_perturbed = [2.0; 2.00001]  # Tiny change!
x_perturbed = inv(A) * b_perturbed
println("\nTiny change in b → HUGE change in x:")
println("New x = ", x_perturbed)
println("Change in x = ", norm(x_perturbed - x_direct))

println("\n" * "="^50 * "\n")

# Method 2: Regularization (Strang's fix!)
σ² = 0.01  # Regularization parameter
println("REGULARIZED solution (minimize ||Ax-b||² + σ²||x||²):")

# This gives (A'A + σ²I)x = A'b
x_reg = (A'*A + σ²*I) \ (A'*b)
println("x = ", x_reg)
println("||x|| = ", norm(x_reg), " ← Reasonable!")

# Small perturbation barely affects regularized solution
x_reg_perturbed = (A'*A + σ²*I) \ (A'*b_perturbed)
println("\nWith same tiny change in b:")
println("New x = ", x_reg_perturbed)
println("Change in x = ", norm(x_reg_perturbed - x_reg), " ← Small!")
```

```julia
function compare_solutions(A, b, σ²)
    U, Σ_vals, V = svd(A)

    println("Original singular values: ", Σ_vals)

    # Without regularization: use 1/σᵢ
    println("\nWithout regularization, we divide by:")
    for σ in Σ_vals
        println("  1/$σ = $(1/σ)")
    end

    # With regularization: use σᵢ/(σᵢ² + σ²)
    println("\nWith regularization (σ² = $σ²), we use:")
    for σ in Σ_vals
        regularized = σ/(σ^2 + σ²)
        println("  $σ/($σ² + $σ²) = $regularized")
    end

    println("\nNotice: small singular values are 'dampened'!")
end

A = [1.0 1.0; 1.0 1.0001]
b = [2.0; 2.0]
compare_solutions(A, b, 0.01)
```

```julia
using LinearAlgebra
using Random

# Define dimensions
rows = 100
cols = 50

# Create a base matrix with random values
Random.seed!(123) # for reproducibility
A = randn(rows, cols)

# Introduce linear dependency
# Make subsequent columns a linear combination of previous columns with small perturbations
for j in 2:cols
    # Make column j a linear combination of columns 1 to j-1
    linear_combination = zeros(rows)
    for i in 1:(j-1)
        linear_combination += randn() * A[:, i] # random weights for linear combination
    end
    # Add a small random perturbation to the linear combination
    A[:, j] = linear_combination + 0.01 * randn(rows) # small perturbation
end

println("Generated a highly dependent matrix A with dimensions: ", size(A))
```

## Implement gram-schmidt

### Subtask:
Write Julia code to perform the Gram-Schmidt orthogonalization process on matrix A.


**Reasoning**:
Implement the Gram-Schmidt orthogonalization process as described in the instructions.



```julia
function gram_schmidt(A::Matrix{Float64})
    m, n = size(A)
    Q = zeros(m, n)

    for j in 1:n
        v = A[:, j] # Start with the original column

        # Subtract the projections onto previously orthogonalized vectors
        for i in 1:(j-1)
            q = Q[:, i]
            v -= dot(v, q) * q
        end

        # Normalize the vector
        Q[:, j] = v / norm(v)
    end

    return Q
end

# Apply the Gram-Schmidt process to the generated matrix A
Q = gram_schmidt(A)

println("Original matrix A (first 5 rows):")
display(A[1:5, :])

println("\nOrthogonalized matrix Q (first 5 rows):")
display(Q[1:5, :])

# Verify orthogonality (Q'Q should be close to identity)
println("\nChecking orthogonality (Q'Q):")
display(Q' * Q)
```

## QR Decomposition Explained

QR decomposition is a way to break down a matrix $A$ into the product of two other matrices, $Q$ and $R$. The decomposition is expressed as:

$A = QR$

where:

*   $A$ is the original matrix.
*   $Q$ is an orthogonal matrix. An orthogonal matrix has the property that its transpose is equal to its inverse, meaning $Q^T Q = I$, where $I$ is the identity matrix. This also implies that the columns of $Q$ are orthonormal vectors (they are mutually orthogonal and each has a length of 1).
*   $R$ is an upper triangular matrix. An upper triangular matrix has all zero entries below the main diagonal.

The QR decomposition is a fundamental tool in linear algebra with various applications, including:

*   **Solving Linear Systems:** For a system $Ax = b$, if $A = QR$, the system becomes $QRx = b$. Since $Q$ is orthogonal, we can multiply both sides by $Q^T$ to get $Q^T QR x = Q^T b$, which simplifies to $Ix = Q^T b$, or $Rx = Q^T b$. This system can be easily solved for $x$ using back substitution because $R$ is upper triangular.
*   **Least Squares Problems:** QR decomposition is a numerically stable method for solving linear least squares problems, which arise in data fitting and regression.
*   **Eigenvalue Computation:** It is used in the QR algorithm for computing the eigenvalues and eigenvectors of a matrix

```julia
function gram_schmidt_qr(A::Matrix{Float64})
    m, n = size(A)
    Q = zeros(m, n)
    R = zeros(n, n)

    for j in 1:n
        v = A[:, j] # Start with the original column

        # Subtract the projections onto previously orthogonalized vectors
        for i in 1:(j-1)
            q = Q[:, i]
            R[i, j] = dot(v, q)
            v -= R[i, j] * q
        end

        # Calculate the norm of the modified vector and normalize
        R[j, j] = norm(v)
        if R[j, j] < 1e-15 # Handle potential linear dependency
            Q[:, j] = zeros(m) # Set to zero vector if norm is close to zero
        else
            Q[:, j] = v / R[j, j]
        end
    end

    return Q, R
end

# Call the gram_schmidt_qr function with matrix A
Q, R = gram_schmidt_qr(A)

# Display the resulting matrix Q
println("\nOrthogonal matrix Q:")
display(Q)
```
